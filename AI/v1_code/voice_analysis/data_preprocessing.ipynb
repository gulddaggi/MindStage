```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def extract_audio_features(wav_path):
    """다양한 음향 특징 추출"""
    
    # librosa로 로드 (자동으로 mono 변환)
    y, sr = librosa.load(wav_path, sr=None)
    
    features = {}
    
    # 1. 기본 정보
    features['duration'] = librosa.get_duration(y=y, sr=sr)
    features['sample_rate'] = sr
    features['num_samples'] = len(y)
    
    # 2. Zero Crossing Rate (음성/음악 구분에 유용)
    zcr = librosa.feature.zero_crossing_rate(y)
    features['zcr_mean'] = zcr.mean()
    features['zcr_std'] = zcr.std()
    
    # 3. Spectral Features
    spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
    features['spectral_centroid_mean'] = spectral_centroids.mean()
    features['spectral_centroid_std'] = spectral_centroids.std()
    
    spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)[0]
    features['spectral_rolloff_mean'] = spectral_rolloff.mean()
    
    # 4. MFCC (Mel-Frequency Cepstral Coefficients) - 음성 인식의 핵심!
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    for i in range(13):
        features[f'mfcc_{i}_mean'] = mfccs[i].mean()
        features[f'mfcc_{i}_std'] = mfccs[i].std()
    
    # 5. Chroma Features (음높이 관련)
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    features['chroma_mean'] = chroma.mean()
    features['chroma_std'] = chroma.std()
    
    # 6. Tempo (BPM)
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
    features['tempo'] = tempo
    
    # 7. RMS Energy
    rms = librosa.feature.rms(y=y)
    features['rms_mean'] = rms.mean()
    features['rms_std'] = rms.std()
    
    # 8. Pitch (F0) - 음성의 기본 주파수
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    pitch_values = []
    for t in range(pitches.shape[1]):
        index = magnitudes[:, t].argmax()
        pitch = pitches[index, t]
        if pitch > 0:
            pitch_values.append(pitch)
    
    if pitch_values:
        features['pitch_mean'] = np.mean(pitch_values)
        features['pitch_std'] = np.std(pitch_values)
        features['pitch_min'] = np.min(pitch_values)
        features['pitch_max'] = np.max(pitch_values)
    
    return features

# 사용
features = extract_audio_features("/home/j-k13b204/S13P31B204/AI/v1_code/voice_analysis/ckmk_a_bm_m_e_3549.wav")
for key, value in features.items():
    print(f"{key}: {value:.4f}" if isinstance(value, float) else f"{key}: {value}")
```

```python
import librosa
import numpy as np

# --- 0. 분석할 파일 경로 ---
filename = '/home/j-k13b204/S13P31B204/015.감성_및_발화_스타일별_음성합성_데이터/01.데이터/2.Validation/원천데이터/VS1/4.감정X발화스타일/12.당황X대화체/0045_G2A3E6S4C0_PMK/0045_G2A3E6S4C0_PMK_001998.wav'

# --- 1. 오디오 로드 ---
try:
    # sr=None: 원본(48000Hz) 유지
    y, sr = librosa.load(filename, sr=None)
    total_duration = librosa.get_duration(y=y, sr=sr)
    print(f"파일 로드 성공: {filename}")
    print(f"파일의 총 길이: {total_duration:.2f}초\n")
except Exception as e:
    print(f"파일 로드 오류: {e}")
    raise SystemExit

# --- 2. 임계값 설정 ---
# 필요 시 튜닝: 현재는 예시 값
silence_threshold = 0.5    # 이 값(초)보다 긴 묵음이 있으면 이상
drop_threshold = -0.10     # 이 값보다 큰 음량 하락(음수, 절댓값 큼)이면 이상

# RMS 계산 설정 (librosa 기본 값과 맞춰줌)
FRAME_LENGTH = 2048
HOP_LENGTH = 512

# --- 3. (회귀 1) 묵음의 길이 측정 ---
print("--- 1. '묵음의 길이' 분석 ---")

# split은 [start_sample, end_sample] (샘플 인덱스)를 반환
intervals = librosa.effects.split(y, top_db=40)

speech_intervals = []
for start_sample, end_sample in intervals:
    start_t = librosa.samples_to_time(start_sample, sr=sr)
    end_t = librosa.samples_to_time(end_sample, sr=sr)
    speech_intervals.append((start_t, end_t))

silence_durations = []

if not speech_intervals:
    # 전체가 묵음
    silence_durations.append(total_duration)
else:
    # 1) 첫 발화 이전 묵음
    first_start = speech_intervals[0][0]
    if first_start > 0.01:
        silence_durations.append(first_start)

    # 2) 발화 사이 묵음
    for (prev_s, prev_e), (next_s, next_e) in zip(speech_intervals, speech_intervals[1:]):
        gap = next_s - prev_e
        if gap > 0.01:
            silence_durations.append(gap)

    # 3) 마지막 발화 이후 묵음
    last_end = speech_intervals[-1][1]
    tail_silence = total_duration - last_end
    if tail_silence > 0.01:
        silence_durations.append(tail_silence)

max_silence = max(silence_durations) if silence_durations else 0.0
print(f"가장 긴 묵음 시간 (회귀 값): {max_silence:.2f}초")

# --- 4. (회귀 2) 급격한 소리 감소 측정 ---
print("\n--- 2. '급격한 소리 감소' 분석 ---")

rms_energy = librosa.feature.rms(
    y=y,
    frame_length=FRAME_LENGTH,
    hop_length=HOP_LENGTH,
    center=True
)[0]

if len(rms_energy) < 2:
    max_volume_drop = 0.0
    drop_time = 0.0
else:
    energy_changes = np.diff(rms_energy)  # frame(t) - frame(t-1)
    max_volume_drop = np.min(energy_changes)  # 가장 큰 하락(가장 음수)
    drop_index = np.argmin(energy_changes) + 1  # 해당 frame index
    drop_time = librosa.frames_to_time(drop_index, sr=sr, hop_length=HOP_LENGTH)

print(f"가장 급격한 소리 감소량 (회귀 값): {max_volume_drop:.4f}")
print(f"발생 시점: {drop_time:.2f}초")

# --- 5. 최종 판단 (분류) ---
print("\n--- 3. 최종 판단 (분류) ---")

is_long_silence = max_silence >= silence_threshold
is_sudden_drop = max_volume_drop <= drop_threshold  # drop_threshold는 음수

if is_long_silence or is_sudden_drop:
    print("[판단: 🚨 이상 탐지]")
    if is_long_silence:
        print(f" - 사유: 가장 긴 묵음({max_silence:.2f}초)이 임계값({silence_threshold:.2f}초)을 초과")
    if is_sudden_drop:
        print(f" - 사유: 가장 급격한 소리 감소({max_volume_drop:.4f})가 임계값({drop_threshold:.4f})보다 급격")
else:
    print("[판단: ✅ 정상]")
    print("묵음의 길이와 소리 감소량 모두 설정한 기준 내에 있음")
```

```python
import librosa
import numpy as np

# ---- 1. 기본 feature 추출 ----
def extract_window_features(y, sr, win_sec=0.5, hop_sec=0.25):
    win_len = int(win_sec * sr)
    hop_len = int(hop_sec * sr)
    n_samples = len(y)
    feats = []

    for start in range(0, n_samples - win_len + 1, hop_len):
        end = start + win_len
        seg = y[start:end]

        rms = float(np.sqrt(np.mean(seg**2)) + 1e-12)
        zcr = float(librosa.feature.zero_crossing_rate(seg)[0].mean())
        centroid = float(librosa.feature.spectral_centroid(y=seg, sr=sr)[0].mean())

        pitches, mags = librosa.piptrack(y=seg, sr=sr)
        pitch_vals = []
        for t in range(pitches.shape[1]):
            idx = mags[:, t].argmax()
            f0 = pitches[idx, t]
            if f0 > 0:
                pitch_vals.append(f0)
        pitch_mean = float(np.mean(pitch_vals)) if pitch_vals else 0.0

        feats.append({
            "start": start / sr,
            "end": end / sr,
            "rms": rms,
            "zcr": zcr,
            "centroid": centroid,
            "pitch_mean": pitch_mean
        })
    return feats


# ---- 2. robust z-score (median / IQR 기반) ----
def robust_z(x):
    x = np.array(x)
    med = np.median(x)
    q1, q3 = np.percentile(x, [25, 75])
    iqr = q3 - q1
    if iqr == 0:
        iqr = 1e-8
    return (x - med) / (iqr + 1e-8)


# ---- 3. 비지도 이상 감지 (값 + 변화량 Δ) ----
def unsupervised_delta_anomaly_detection(
    wav_path,
    win_sec=0.5,
    hop_sec=0.25,
    anomaly_percentile=97.0
):
    y, sr = librosa.load(wav_path, sr=None)
    feats = extract_window_features(y, sr, win_sec, hop_sec)

    # --- feature 값 배열화 ---
    rms_vals = np.array([f["rms"] for f in feats])
    zcr_vals = np.array([f["zcr"] for f in feats])
    cent_vals = np.array([f["centroid"] for f in feats])
    pitch_vals = np.array([f["pitch_mean"] for f in feats])

    # --- 변화량(Δfeature): 프레임 간 절대 변화 ---
    def delta(x): return np.concatenate([[0], np.abs(np.diff(x))])
    rms_delta = delta(rms_vals)
    pitch_delta = delta(pitch_vals)
    cent_delta = delta(cent_vals)

    # --- robust scaling ---
    z_rms = np.abs(robust_z(rms_vals))
    z_pitch = np.abs(robust_z(pitch_vals))
    z_cent = np.abs(robust_z(cent_vals))
    z_delta_rms = np.abs(robust_z(rms_delta))
    z_delta_pitch = np.abs(robust_z(pitch_delta))
    z_delta_cent = np.abs(robust_z(cent_delta))

    # --- anomaly score (값 + 변화량) ---
    scores = (
        0.5 * (z_rms + z_pitch + z_cent) +  # 현재 값이 튀는 정도
        0.5 * (z_delta_rms + z_delta_pitch + z_delta_cent)  # 변화량이 급격한 정도
    )

    # --- 상위 퍼센타일 이상을 anomaly로 판단 ---
    thr = np.percentile(scores, anomaly_percentile)

    anomalies = []
    for f, s in zip(feats, scores):
        if s >= thr:
            anomalies.append({
                "start": round(f["start"], 3),
                "end": round(f["end"], 3),
                "score": float(round(s, 3))
            })

    anomalies = merge_contiguous(anomalies)
    return feats, anomalies, thr


# ---- 4. 인접 구간 병합 ----
def merge_contiguous(anoms, gap_tol=0.0):
    if not anoms:
        return []
    anoms = sorted(anoms, key=lambda x: x["start"])
    merged = [dict(anoms[0])]
    for a in anoms[1:]:
        last = merged[-1]
        if a["start"] <= last["end"] + gap_tol:
            last["end"] = max(last["end"], a["end"])
            last["score"] = max(last["score"], a["score"])
        else:
            merged.append(dict(a))
    for m in merged:
        m["duration"] = round(m["end"] - m["start"], 3)
    return merged


# ---- 5. 실행 ----
if __name__ == "__main__":
    wav_path = "/home/j-k13b204/S13P31B204/AI/v1_code/voice_analysis/ce33dad0-b9bb-401a-ab4b-a2221107b070_interview_full.wav"
    feats, anomalies, thr = unsupervised_delta_anomaly_detection(wav_path, anomaly_percentile=97.0)

    print(f"threshold(score) = {thr:.3f}")
    print("=== Δ-feature 기반 비지도 이상 구간 ===")
    for a in anomalies:
        print(f"{a['start']}s ~ {a['end']}s (dur={a['duration']}s, score={a['score']})")
```
