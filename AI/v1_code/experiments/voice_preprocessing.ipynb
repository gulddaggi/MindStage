```python
# =====================================================
#  Voice EDA (librosa + soundfile)
# =====================================================
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# ===== 파일 경로 =====
path = "/home/j-k13b204/S13P31B204/model_test/Code/voice_analysis/KETI데이터/2-12.감정_분류를_위한_대화_음성_데이터셋/4차_wav/5e2ac3d55807b852d9e01fd6.wav"

# ===== 1) Load =====
y, sr = sf.read(path, dtype="float32")
if y.ndim == 2:
    y = y.mean(axis=1)  # stereo → mono

# ===== 2) Waveform ===== 파형 : 시간에 따른 진폭 변화 그래프
t = np.arange(len(y)) / sr
plt.figure(figsize=(12,3))
plt.plot(t, y)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()

# ===== 3) Spectrogram (STFT) ===== 스펙트로그램 : 시간 + 주파수 + 세기 3차원을 2D로 표현한 것.
S = np.abs(librosa.stft(y, n_fft=1024, hop_length=256))**2
Sdb = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(12,4))
librosa.display.specshow(Sdb, sr=sr, x_axis="time", y_axis="log")
plt.colorbar(label="dB")
plt.title("Spectrogram (dB)")
plt.tight_layout()
plt.show()

# ===== 4) Mel-Spectrogram ===== : 인간의 청각 특성(저음은 민감, 고음은 둔감)을 반영한 스펙트로그램의 변형판.
Mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=64, fmax=8000)
Mel_db = librosa.power_to_db(Mel, ref=np.max)
plt.figure(figsize=(12,4))
librosa.display.specshow(Mel_db, sr=sr, x_axis="time", y_axis="mel")
plt.colorbar(label="dB")
plt.title("Mel-Spectrogram (dB)")
plt.tight_layout()
plt.show()

# ===== 5) MFCC ===== : 멜 스펙트로그램에서 음색·발음 습관을 요약한 벡터.
MFCC = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
plt.figure(figsize=(12,4))
librosa.display.specshow(MFCC, sr=sr, x_axis="time")
plt.colorbar(label="MFCC coeff")
plt.title("MFCC")
plt.tight_layout()
plt.show()

# ===== 6) RMS Energy ===== : 시간에 따른 음성의 에너지(세기) 변화 그래프.
rms = librosa.feature.rms(y=y)[0]
times = librosa.times_like(rms, sr=sr)
plt.figure(figsize=(12,3))
plt.plot(times, rms)
plt.title("RMS Energy")
plt.xlabel("Time (s)")
plt.tight_layout()
plt.show()

# ===== 7) Pitch (F0) with pyin ===== : 사람의 음 높이(목소리의 톤) 를 나타내는 값.
f0, voiced_flag, voiced_prob = librosa.pyin(y, fmin=50, fmax=400, sr=sr)
t_f0 = librosa.times_like(f0, sr=sr)
plt.figure(figsize=(12,3))
plt.plot(t_f0, f0, label="F0")
plt.title("Pitch (F0) over Time")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.tight_layout()
plt.show()

# ===== 8) 기본 통계 =====
print(f"샘플레이트: {sr} Hz")
print(f"총 길이: {len(y)/sr:.2f} 초")
print(f"RMS 평균: {np.mean(rms):.4f}, RMS 표준편차: {np.std(rms):.4f}")
print(f"F0 평균(유효구간): {np.nanmean(f0):.2f} Hz, 표준편차: {np.nanstd(f0):.2f} Hz")
```
